# ライブラリインポート

```python
import cv2
import matplotlib.pyplot as plt
import glob
```

# フォルダ内の画像のパスを取得

```python
files = glob.glob("/content/drive/MyDrive/Colab Notebooks/亀岡さん/*.tif")
for file in files:
    print(file)
print('ファイル数 = {}'.format(len(files)))
```

# 各画像の読み込み

```python
# 画像読み込み
# img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/亀岡さん/愛媛県_EPSG6672_.tif")
img = cv2.imread(files[0])
```

```python
plt.imshow(img)
plt.show()
height, width, channels = img.shape[:3]
print("width: " + str(width))
print("height: " + str(height))
```

# 各画像の切り抜き

```python
# img[top : bottom, left : right]
# サンプル1の切り出し、保存
img1 = img[0 : 1500, 0: 1500]
plt.imshow(img1)
plt.show()
height, width, channels = img1.shape[:3]
print("width: " + str(width))
print("height: " + str(height))
cv2.imwrite("out_sample1.jpg", img1)
```

```python
# img[top : bottom, left : right]
# サンプル2の切り出し、保存
img2 = img[1500 : 3000, 1500: 3000]
plt.imshow(img2)
plt.show()
height, width, channels = img2.shape[:3]
print("width: " + str(width))
print("height: " + str(height))
cv2.imwrite("out_sample2.jpg", img2)
```
